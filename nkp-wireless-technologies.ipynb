###### 1. 802.11 Standards
***

Earlier, we touched on some of the basic characteristics of wireless communications as defined by the 802.11 specification. But that was quite some time ago, and there have been a number of enhancements and refinements to that specification since. So in this presentation, we'll take a look at several of them, beginning with 802.11a.

<a id="wirelessTechnologies180211Standards80211a"></a>
***
###### 802.11a
***

[802.11a.] Now this was created as an extension to the original 802.11 standard and supports a higher bandwidth, up to 54 megabits per second and a frequency spectrum around 5 gigahertz. Now that higher frequency does offer a better quality signal, but by nature, higher frequencies do not travel as far as lower ones do. So it shortens the range of a networks, and further, the higher frequency makes the a signals more difficult to penetrate obstructions, such as walls. Now that's not necessarily a downside, if you will. Sometimes you might actually want to restrict the network a little bit more, condense it, if you will, or confine it to a shorter range. And if it's more difficult to get through things like walls, that might actually provide a little more security.

So again, depending on your perspective, it might not really be a downside. So in terms of the pros, it does offer that higher speed and the use of regulated frequencies, meaning that other devices do not use this frequency. It has effectively been reserved, so this prevents signal interference. 

But on the downside, it can cost more and again, can shorten the range and be more easily obstructed. So if you are looking for a wider range that penetrates more easily, then yes, this is certainly a con. 

< [Table of Contents](#top) | [References](#references) >
<a id="wirelessTechnologies180211Standards80211b"></a>
***
###### 802.11b
***

[802.11b.] Now b, supports bandwidths of up to 11 megabits per second. So considerably slower, this is comparable to the traditional Ethernet around 10 megabits per second. And that, in and of itself, is quite old these days. And it also uses unregulated 2.4 gigahertz signaling frequency, meaning a lot of vendors, a lot of devices use the same frequency. But since it is very commonly produced, this can lower the production costs. So ultimately, it can incur interference from devices using the same 2.4 gigahertz range. Some common examples are microwaves ovens, cordless phones, and various other appliances. So they can disrupt the signals.

So in terms of the pros, it is a lower cost, and it has that better range. And it's not as easily obstructed.

But on the cons side, it is the slowest maximum speed. And again, susceptible to interference because of that unregulated frequency band. So you might think well, if it's not as fast as ``a``, and it suffers from possible interference, and simply the fact that it uses the letter ``b``, you might think that it came out after a. In fact, a and b were released at the same time, and it really just comes down to what you prefer. If you want the lower cost and the better range, you went with b. But there was the tradeoff in that it wasn't as fast. But a was faster, but more easily obstructed and things along those lines, didn't travel as far. So there was always the tradeoff in either case.

###### 802.11g
***

[802.11g.] So along came g, which emerged in 2002 and 2003, and this combines the best of both ``a`` and ``b``. It supports the higher bandwidth of up to 54 megabits per second, but uses the 2.4 gigahertz frequency for greater range. This was also backward compatible, in other words, it worked with the already in place b wireless adapters that you might have had. So we don't have to change out all of our adapters to take advantage of the higher speed.

In terms of the pros, you get that fast maximum speed, again up to 54 megabits per second. You get good signal range, which is not easily obstructed, but it could have been more expensive than the b that was already implemented. Plus, since you did already implement the b, you are looking at basically junking all of that and incurring the extra cost of a new infrastructure. Plus, it still might suffer from the interference, because it was using 2.4 gigahertz.

###### 802.11n
***

[802.11n.] So then n came along. This was often referred to as just Wireless N. But this improves on ``g`` by offering greater bandwidth and utilizing multiple wireless signals and antennas instead of just one. This is known as MIMO technology, which we will actually talk about later on. Very quick and short, it is multiple antennas as opposed to just one. A lot of the early wireless routers had a single antenna. So this can support bandwidths of up to 300 megabits per second and a better range and increased signal intensity.

So in terms of the pros, we now have the fastest maximum speed, 300 megabits per second, the best signal range, and the most resistant to signal interference, and it's still backward compatible. Cons, the standard is not yet finalized, so it may affect things moving forward. And again, it's more expensive than 802.11g.

###### 802.11ac
***

[802.11ac.] Then we have 802.11ac, which is the newest generation of Wi-Fi signaling. This utilizes dual band wireless technology, which means it supports simultaneous connections over both 2.4 and 5 gigahertz. It is backward compatible with ``802.11b``, ``g``, and ``n``. And supports bandwidths that have been rated up to 1,300 megabits per second on 5 gigahertz and 450 megabits per second on 2.4. So significantly higher, in fact in both cases, using either frequency. So again, this really pulls together the best of everything, but this can certainly be a little more expensive as well, and does represent yet another upgrade. So if you have already implemented your existing infrastructure with something else, once again, you are looking at another upgrade. So there certainly can be those downsides as well. Ultimately, you just want to try to ensure that you have good range, good signal quality, good speed, and of course, security, but at the same time, while not having to continually upgrade and incur additional costs.

###### 2. Cellular Technologies

###### GSM
***

GSM - Global System for Mobile communication. This is a digital mobile telephony system that currently represents over 82% [precisely 82.4%.] of all global mobile connections. Now, it operates at one of two frequencies, 900 or 1800 MHz. And, of course, fundamentally, just digitizes and compresses data prior to sending it down a channel. But one of the inherent challenges is, of course, the tremendous number of people using mobile communications these days. There are billions worldwide. ###### Time Division Multiple Access - TDMA
***

[TDMA.] And one approach is what's known as TDMA, Time Division Multiple Access. Again, multiple people using the same media at the same time. So this is a technology used in digital cellular communication, whereby the channels are divided into three time slots. This increases the amount of data that can be carried, essentially, threefold. Now, this is used by Digital-Advanced Mobile Phone Services, or D-AMPS, GSM [Global System for Mobile communications.] that we just talked about, and Personal Digital Cellular, [abbreviated as PDC.] those are just different types. And D-AMPS and PDC, in essence, make up the remaining percentage of communications. 

But the idea behind TDMA is that each channel, and there are a tremendous number of channels available, is divided by time. And then each user or each device synchronizes to that time. So when I'm communicating, I use time slot A. When you are communicating, you use time slot B, and someone else uses C. And we only pay attention, if you will, to the data in our specific time slots. So I don't hear your conversation, you don't hear mine, because of the difference in the time slots. Now this does require some pretty specific synchronization, but voice data is not all that much. So, in other words, it's not a big deal to break it up into these smaller chunks, if you will. And it accommodates multiple users on the same channel. So, again, we can increase the overall throughput three times per channel.

< [Table of Contents](#top) | [References](#references) >
<a id="wirelessTechnologies2CellularTechnologiesCodeDivisionMultipleAccessCDMA"></a>
***
###### Code Division Multiple Access - CDMA
***

[CDMA.] Now another method of accommodating that multiple access is CDMA, or Code Division Multiple Access. This is used in second-generation, 2G, and third-generation, 3G, wireless communications. And I'm certain, if you have a mobile device, you are very familiar with what we call 3G. So this is just another form of multiplexing which allows numerous signals on a single transmission channel. And this optimizes the available bandwidth. It's commonly used in ultra-high frequency cellular systems running at either 800 MHz and 1.9 GHz. But the idea is that each transmission has a specific code attached to it. And your device is programmed, for lack of a better word, to only pay attention to the specific code that matches your connection. So again, my conversation uses Code 1. Your conversation uses Code 2. We only pay attention to the appropriate code. So, again, I do not hear your conversation, you do not hear mine. So, this way, again, we can facilitate multiple access over the same channel. 

So in terms of the benefits, this offers good quality signal, there's no division in terms of the time, so the signal quality is a little better. It offers flexible allocation of resources. It's easy scalable because you can just add more codes. And it's compatible with other cellular technologies. 

Disadvantages, as the network grows, overall quality of service decreases. If you end up with too many codes, that's a lot of processing. So the more devices, the more codes you need to maintain and manage. So, in essence, it can become a bottleneck. But ultimately, that is how, these days, we facilitate the multiple access requirements of cellular communications.

The length is the distance between the 2 peaks or the 2 troughs, for example.

###### 5. Channel Bandwidth
***

In this presentation, we'll take a closer look at channel bandwidth. And this is something I referred to earlier, when I mentioned how channels more or less equate to lanes on a highway, and the more lanes you have, the more traffic you can handle. So in terms of data rates, your speed, can be directly affected by the channel bandwidth of a wireless signal. 

Now we see two examples here, 802.11n using 20 or 40 MHz channels and 802.11ac using 20,40, 80, and 160 MHz channels. Now, n uses the core frequency of 2.4 GHz. And ac, the core frequency of 5 GHz. But the question I get quite often is, well, what are these two hertz values? I thought the transmission was 2.4 GHz or 5 GHz depending on your implementation. Why am I seeing this megahertz values on the channels?

Well, that comes down to the basic fact that anything measured in hertz, recall, is simply a measure of how often something happens.

So the core frequencies 2.4 GHz and 5 GHz are certainly correct. You send out a signal using that frequency. That refers to the waves, how often those waves happen in one second. But then you have to consider how often you can send that signal. So let me give you a simpler example.

If you just speak a word, let's say the word hello, that word produces sound waves, and those sound waves come out at a specific frequency. It doesn't what it is, let's just call it frequency x. So you say the word hello, you just sent a signal, and the sound waves came out at frequency x. But you can, of course, say that word again and again and again and again. So how many times you can say that word within one second refers to another measure in hertz, how many times can you send that information? So that's where the channel comes in. The more often you can say it, the more information you are transmitting. So that's where the channels allow for x amount of data depending on how often they can support that core transmission.

So, again, the signal itself, the core data, is being sent out at, let's say 2.4 GHz. So the channels then, are how many times you can send that transmission across that channel. So those are the two separate hertz values. And clearly, the more megahertz you have per channel, the more times, if you will, you can send that signal. So you end up with higher bandwidth overall.

###### 6. Channel Bonding
***

Now, in this presentation, we'll take a look at what's known as channel bonding. [Channel Bonding.] And this is commonly used in IEEE 802.11, or wireless communications. And is the process of taking two adjacent channels using different frequencies and combining them. This, of course, increases the overall throughput between those devices. 

Now, we actually touched on this earlier when we looked at the various specifications of 802.11, the a, b, g, n, and ac, and we saw various throughputs, of course. But for the most part, if we saw a throughput of x on a 20 MHz channel, then, of course, you could get a throughput of 2x on a 40 MHz channel. So those widths of 20 MHz per channel were established quite sometime ago in wireless communications. But there's nothing that prevents us from taking two of them and putting them together and creating a 40 MHz width. But of course, what happens here, let's just go with very easy numbers and let's imagine we only had two available to begin with, then clearly, you are reducing the number of channels.

If you take those two and put them together to create one, well, now you only have one channel. So you will inherently have fewer channels when you bond. But if you aren't concerned with the number of channels and are really only concerned with the throughput, then, of course, bonding is the way to go. 

Now, this was very popular in Wi-Fi implementations and remains so as it provides additional functionality for what you can support. And, of course, that overall increased throughput, and this was initially released with 802.11n.

###### 7. MIMO/MU-MIMO
***

In this presentation we'll take a look at MIMO or Multiple Input Multiple Output. 

< [Table of Contents](#top) | [References](#references) >
<a id="wirelessTechnologies7MIMOMUMIMOMIMO"></a>
***
###### Multiple Input Multiple Output - MIMO
***

[MIMO.] And this is actually an antenna technology for wireless communications whereby you can send multiple signals at once using an array of antennas. Now, these multiple antennas are combined at each end of the circuit. And with respect to each end, it doesn't refer to the fact that there is one on the left and one on the right of something like a wireless router, but rather multiple antennas at the transmitter, multiple antennas at the receiver. So by using these multiple signals you can minimize errors and optimize your data speed. Now this is supported by 802.11n and ac. 

And fundamentally, it's taking into consideration the fact that electromagnetic signals bounce off walls and other obstructions and hence can take different paths from device to device. They are quite literally echoing if you will. So this causes delays and increased errors with respect to any given transmission. So by using multiple signals over those multiple antennas you get greater reliability and greater throughput overall.

Now it should also be mentioned that what we are seeing in the graphic here is what's known as single session or single user MIMO. It does not mean that you cannot have multiple users connected. Since the earliest days of wireless networking you, of course, could have more than one user connected. But it just means that it kind of has to deal with each session, one at a time. So as far as the router is concerned, for example, it sees sessions 1 through 3. And you can imagine this as trying to have different conversations with three different people. You can do it, but you have to move from person to person to person. So each person gets their own turn, if you will, with respect to the communication, so clearly it's not as fast.

###### Multi-User MIMO
***

[MU-MIMO.] So Multi-User MIMO was introduced to, in essence, address that issue. This is an enhanced form of MIMO. And it enables multiple independent radio terminals to access a system. So, it uses various algorithms and various methods of multiplexing to effectively allow simultaneous communication with multiple users. And you can see it can actually still use both. It can use the single session capabilities if, and/or when necessary so we see sections 1 and 2 actually being dealt with separately.

Then we see 3 and 4 are basically on their own, if you will. So 1, 2 are combined. Then 3 and 4 are independent. But all of this is happening at the same time. So, with this now we've got multiple radio terminals, they can provide significant performance enhancement. Because we now have these multiple users simultaneously accessing the same channel. So with this we can get significantly higher data rates. For example in 802.11ac, the increased theoretical maximum speed is anywhere from 3.47 gigabits per second up to 6.93. Now you probably won't see those in practical implementations. Those are laboratory tests if you will where everything is perfect. But clearly with this multi-user capability and the ability to have these simultaneous conversations if you will, the overall throughput can be dramatically increased.
